<strong>Selección de características</strong>
<p>La selección de funciones es una de las tareas más importantes en el aprendizaje automático. Es especialmente útil cuando se trata de datos de gran dimensión o cuando su conjunto de datos contiene una gran cantidad de características y un número limitado de observaciones. La reducción de características también ahorra tiempo de almacenamiento y computación y hace que sus resultados sean más fáciles de entender.
Las técnicas comunes de selección de características incluyen:</p>

<ul>
    <li>Regresión paso a paso: agregar o eliminar funciones secuencialmente hasta que no haya una mejora en la precisión de la predicción.</li>
    <li>Selección secuencial de características: Agregue o elimine iterativamente variables de predicción y evalúe el efecto de cada cambio en el rendimiento del modelo.</li>
    <li>Regularización: utilizar estimadores de contracción para eliminar las características redundantes al reducir sus ponderaciones (coeficientes) a cero</li>
    <li>Análisis de componentes de vecindad (NCA): determinar el peso que cada característica tiene para predecir la salida, de modo que las características con menos peso se puedan descartar.</li>
</ul>

## Regresión Logistica

<p>Regresión logistica. Ajusta un $x_2$ modelo que puede predecir la porbabilidad de que una respuesta pertenenzca a una clase o a otra (clasificacion binaria). Es usada para inicialisar un problema de clasificación binario. Se usa cuando los datos pueden separarse por una simple barrera lineal, es una base para aplicar metodos de clasificación más complejos</p>
        
      
 <p> Es un metodo clásico, utiliza la función de pérdida $log(1-e^{-yf(x)})$ minimizamos</p>
       
 $$\min\limits_{moldels\;f}\frac{1}{n}\sum_{i=1}^n log(1+e^{-y_if(x_i)})$$
 <p>
  \begin{equation}
  \label{foo}
  f(\bar{x}_i)=\sum_{j=1}^p \beta_jx_{ij}
  \end{equation}
   </p>
 <p>Así se tiene que la regresion logistica es 
   <p>
  \begin{equation}
    \label{reglogidtica}
   \min\limits_{\;\beta_1\ldots\beta_p}\frac{1}{n}\sum_{i=1}^n log\left(1+e^{-y_i\sum\limits_{j=1}^p \beta_jx_{ij}}\right)
       \end{equation}
     </p>
     
    
<p> La función logistica inventada por verhultz </p>

\begin{equation}
f(t)=\frac{e^t}{1+e^t}
\end{equation}


\begin{equation}
\mathbb{P}(Y_i=1|\bar x_i,\beta)=\frac{e^{\bar x_i\beta}}{1+e^{\bar x_i\beta}}
\end{equation}

<p>donde $\bar x_i\beta=\sum\limits_{j=1}^p x_{ij}\beta_j$</p>

\begin{equation}
\mathbb{P}(Y_i=-1|\bar x_i,\beta)=1-\frac{e^{\bar x_i\beta}}{1+e^{\bar x_i\beta}}=\frac{1}{1+e^{\bar x_i\beta}}
\end{equation}


<p>La minimizaciñon se puede interpretar como un problema de maxima verisimilitud, bueno solo verosimilitud, de esa manera se puede demostrar la forma de la funcióna minimizar. Los pasos para la regresión logistica ze pueden resumir en lso siguientes</p>

<div class="algoritmo">
    <strong>Algoritmo: Regresión logística</strong>
<ul class="nobullet">
    <li><p><strong>Paso 1.</strong> Separar los datos aleatoriamente para entrenamiento y para prueba </p></li>
    <li><p><strong>Paso 2.</strong> Estimación de los coeficientes (entrenamiento del modelo) 
    
    $$\hat{\beta}=\min\limits_{\;\beta_1\ldots\beta_p}\frac{1}{n}\sum_{i=1}^n log\left(1+e^{-y_i\sum\limits_{j=1}^p \beta_jx_{ij}}\right)$$
    
</p></li>
<li><p><strong>Paso 3.</strong> Califica el modelo, para cada $\bar x_i$ en el conjunto de prueba
    
    $$f(\bar{x}_i)=\sum_{j=1}^p \hat{\beta}_jx_{ij}$$
 </p></li>
 <li><p><strong>Paso 4.</strong> Se evalúa el modelo</p></li>
</ul>
