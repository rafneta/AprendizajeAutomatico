<strong>Mejoramiento de modelos</strong>
</p>
<p>Mejorar un modelo significa aumentar su precisión y poder predictivo y evitar el sobreajuste (cuando el modelo no puede distinguir entre datos y ruido). La mejora del modelo involucra ingeniería de características (selección y transformación de características) y ajuste de hiperparámetros.</p>

<ul>
    <li>Selección de características: identificación de las características más relevantes, o variables, que proporcionan la mejor capacidad de predicción al modelar sus datos. Esto podría significar agregar variables al modelo o eliminar variables que no mejoran el rendimiento del modelo.</li>
    <li>Transformación de características: convierte características existentes en nuevas características utilizando técnicas como el análisis de componentes principales, la factorización de matrices no negativas y el análisis de factores.</li>
    <li>Ajuste de hiperparámetros: el proceso de identificación del conjunto de parámetros que proporciona el mejor modelo. Los hiperparámetros controlan cómo un algoritmo de aprendizaje automático se ajusta al modelo de los datos.</li>
</ul>

<strong>Selección de características</strong>
<p>La selección de funciones es una de las tareas más importantes en el aprendizaje automático. Es especialmente útil cuando se trata de datos de gran dimensión o cuando su conjunto de datos contiene una gran cantidad de características y un número limitado de observaciones. La reducción de características también ahorra tiempo de almacenamiento y computación y hace que sus resultados sean más fáciles de entender.
Las técnicas comunes de selección de características incluyen:</p>

<ul>
    <li>Regresión paso a paso: agregar o eliminar funciones secuencialmente hasta que no haya una mejora en la precisión de la predicción.</li>
    <li>Selección secuencial de características: Agregue o elimine iterativamente variables de predicción y evalúe el efecto de cada cambio en el rendimiento del modelo.</li>
    <li>Regularización: utilizar estimadores de contracción para eliminar las características redundantes al reducir sus ponderaciones (coeficientes) a cero</li>
    <li>Análisis de componentes de vecindad (NCA): determinar el peso que cada característica tiene para predecir la salida, de modo que las características con menos peso se puedan descartar.</li>
</ul>

<strong>Transformación de características</strong>

<p>La transformación de características es una forma de reducción de dimensionalidad. Como vimos en la sección 3, las tres técnicas de reducción de dimensionalidad más comúnmente utilizadas son:</p>

<ul>
    <li>Analisis de Componentes prncipales. Realiza una transformacion lineal de los datos donde la mayor varianza o informacion de todas las variables es capturada por sus principales componentes. Esta importacua es jerarquizada por la primera variables la segunda, esto lo he hecho en econometria. Cad componente pricipal es una combinacion lineal de las primeras variables donde no hay información redundante pues son ortoganles entre si </li>
    <li>analissi de Factor. Identifica la correlació entre las variables para proveer de una representacion en terminos de una numero peueño de factors comunes o latentes. Puede que tengas varibles que dependan entre si ( se traslapen) esta tecnica te permite obtener un modelo que estima el orden de dependencia, piensas que estas varibales dependen de diferentes factores por lo que extraes estos factores y expresas las variables como una combinación lineal de ellos</li>
    <li>Factorizaciñon d mAtrices no NEgativa. Usada cuando los terminos del modelo deben epresentacrse en cantidades no negativas como cantidades fisicas. Esta tecnica esta basada en un esacio de aproximación de rango bajo.</li>
</ul>

<strong>Ajuste de hiperparámetros</strong>

<p>Al igual que muchas tareas de aprendizaje automático, el ajuste de parámetros es un proceso iterativo. Comienza por establecer los parámetros según una "mejor estimación" del resultado. Su objetivo es encontrar los valores "mejores posibles", los que producen el mejor modelo. A medida que ajusta los parámetros y el rendimiento del modelo comienza a mejorar, puede ver qué parámetros son efectivos y cuáles requieren sintonía.
Tres métodos comunes de ajuste de parámetros son:
<br>• Optimización bayesiana
<br>• Búsqueda de cuadrícula
<br>• Optimización basada en gradientes</p>

<p>En el análisis de conglomerados, los datos se dividen en grupos según alguna medida de similitud o característica compartida.         <li>K-Medoids. similar a K-Means, pero el centro de los clusters coincide con los puntos (datos). Se usa cuando el numro de clusters es conocido, para el clustering rápido e datos categoricos, para escalar a grandes cantidades de datos</li>
        <li>CLustering Jerárquico. Produce conjuntos anidados de clusters mediante el analisis de similitudes entre  pares de puntos y agrupando objetos d entro de un arbol binario jerarquico. Es usado cuando no se sabe que cantidad de clusters tendr´s en tus datos, cuando se quiere una guía de vsualización de su selección. El dendrograma parece como un organizagrama</li>
        <li>Mapas-Autoorganziados. Redes neuronales basados en clustering que transforman un conjunto de datos en un mapeo 2D que presrva la topología. Es usado para visualizar datos de alta dimensión 2D y 3D, para reducir la dimensión de los datos presevando su topología (forma)</li>
