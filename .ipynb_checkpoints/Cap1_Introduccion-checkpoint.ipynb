<p>Dada una estrada esta se le debe asociar una determinada salida. POr ejemplo el porblema de <strong>clasifiación:</strong> Clasificar tipos de imagenes, identificar cuentas de redes sociales aprocrifas, reconosimiento de digitos escritos a mano </p>

<p>Aqui se tienen una cantidad de datos que sirven de entrenamiento, el usuario provee de las salidas correctas (etiquetas) </p>

<p>Los datos se deben de separa en datos de entrenamiento y datos de prueba</p>

<p>El comportamiento de la predicción puede ser diferente de los datos de entrenamiento respectoa laso atos de prueba, esto se conoce como overfitting</p>

## Turting test

<p>Esta actividad puede ser asignada como para diseñar una tarea. </p>


<p>En la prueba, un interrogador humano interactúa con dos jugadores, A y B, al intercambiar mensajes escritos (en un chat). Si el interrogador no puede determinar qué jugador, A o B, es una computadora y cuál es un humano, se dice que la computadora pasa la prueba. El argumento es que si una computadora es indistinguible de un ser humano en una conversación general en lenguaje natural, entonces debe haber alcanzado la inteligencia a nivel humano</p>
